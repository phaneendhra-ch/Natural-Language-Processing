# Introdction to Spacy

### Workflow of Spacy Library

#### Source Link : https://spacy.io/usage/linguistic-features

```python
import spacy
```

```python
# Load English Core Language of Smaller version
# This can be downloaded in the environment using ' python -m spacy download en' 

nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(u'Tesla is looking at buying U.S. startup for ') #Unicode String
```

```python
#Pipeline of Spacy Library
nlp.pipeline
```

```python
nlp.pipe_names
```

```python
#Can extract text and part of speech of each token
for token in doc:
    print(token.text,token.pos_)
```
